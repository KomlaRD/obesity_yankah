```python
# Load libraries
import pandas as pd
from pca import pca
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("dummy_obesity.csv")
df.head()
```

```python
df.dropna(inplace=True)
```

```python
model = pca(normalize=True)
```

```python
result = model.fit_transform(df)
```

```python
model.plot()
```

```python
model.biplot(n_feat=10)
```
